```python
# 11.Write a Python program to get the largest number from a numeric list.

numericList = [1,1,2,3,5,13,21,34,55,89];
largestNumber = []
for item in numericList:
    if(item > largestNumber):
        largestNumber.append(item)
print(largestNumber)
```
